#  ROC Curve (ROC-AUC) for your logistic regression

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
# Get predicted probabilities for the positive class (class 1)
y_probs = lg.predict_proba(pca_test)[:, 1]

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(ytest, y_probs)

# Calculate ROC-AUC score
roc_auc = roc_auc_score(ytest, y_probs)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

print("ROC-AUC:", roc_auc)
```

# Tumor Segmentation

```python
# Load the original 'yes' tumor image (specify the image file you want to process)
image_path = '/content/drive/MyDrive/projectbraincancer/dataset/yes/y10.jpg'
image = cv2.imread(image_path)

# Load the binary mask (segmentation result)
mask_path = '/content/drive/MyDrive/projectbraincancer/dataset/yes/y10.jpg'  # Update with the correct mask image
mask = cv2.imread(mask_path, 0)

# Find contours of the tumor region
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# If there are contours, draw a bounding box around the largest one
if len(contours) > 0:
    # Find the largest contour by area
    largest_contour = max(contours, key=cv2.contourArea)

    # Get the coordinates of the bounding box
    x, y, w, h = cv2.boundingRect(largest_contour)

    # Reduce the size of the bounding box (scale to 80% of original dimensions)
    scale_factor = 0.5
    w = int(w * scale_factor)
    h = int(h * scale_factor)



    # Draw a colored bounding box on the original image
    colored_image = image.copy()
    cv2.rectangle(colored_image, (x, y), (x + w, y + h), (0, 255, 0), 2)  # Green box

    # Display the image with the colored bounding box using Matplotlib
    plt.figure(figsize=(8, 6))
    plt.imshow(cv2.cvtColor(colored_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
    plt.axis('off')
    plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Dropout, Conv2DTranspose, concatenate
from tensorflow.keras.models import Model
```

```python
# Define U-Net model
def unet_model(input_size=(256, 256, 3)):
    inputs = tf.keras.Input(input_size)
    conv1 = tf.keras.layers.Conv2D(64, 3, activation='relu', padding='same')(inputs)
    conv1 = tf.keras.layers.Conv2D(64, 3, activation='relu', padding='same')(conv1)
    pool1 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(conv1)

    # Add more convolutional and pooling layers as needed

    up9 = tf.keras.layers.Conv2DTranspose(64, 2, strides=(2, 2), padding='same')(conv1)
    merge9 = tf.keras.layers.Concatenate()([up9, conv1])
    conv9 = tf.keras.layers.Conv2D(64, 3, activation='relu', padding='same')(merge9)
    conv9 = tf.keras.layers.Conv2D(64, 3, activation='relu', padding='same')(conv9)

    outputs = tf.keras.layers.Conv2D(1, 1, activation='sigmoid')(conv9)

    model = tf.keras.Model(inputs, outputs)
    return model

# Load the image you want to segment
image_path = '/content/drive/MyDrive/projectbraincancer/dataset/yes/y10.jpg'
image = cv2.imread(image_path)
image = cv2.resize(image, (256, 256))  # Resize to match the model's input size
image = np.expand_dims(image, axis=0)  # Add batch dimension

# Load the trained U-Net model (make sure the model is already trained)
model = unet_model()  # Define the U-Net model as provided

# Load the model weights
model.load_weights('unet_weights.h5')  # Provide the path to your model weights

# Perform inference on the image for tumor segmentation
predicted_mask = model.predict(image)

# Adjust the threshold for binary mask creation
threshold = 0.5  # Experiment with different threshold values
binary_mask = (predicted_mask > threshold).astype(np.uint8)

# Display the original image
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image[0], cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Display the segmented tumor region using contours
contours, _ = cv2.findContours(binary_mask[0], cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
segmented_image = cv2.drawContours(image[0], contours, -1, (0, 255, 0), 2)
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(segmented_image, cv2.COLOR_BGR2RGB))
plt.title('Segmented Tumor Region')
plt.axis('off')

plt.show()
```
